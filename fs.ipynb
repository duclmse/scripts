```python
import sys
import shutil
from datetime import datetime, date
from pathlib import Path
```

```python
time_str = '2025-11-20'
time = datetime.strptime(time_str, '%Y-%m-%d')
```

```python
print(time.date())
time.date() == datetime.now().date()
```

```python
path = Path.home() / "Downloads/logs/20251119.cmm.zip"
if not path.exists():
    raise FileNotFoundError(f"Directory {path} does not exist")

matching_files = []

for file in path.glob(f"cmm*.gz"):
    if not file.is_file():
        continue

    # Get file modification time
    ts = file.stat().st_mtime
    file_mtime = datetime.fromtimestamp(ts).date()
    print(f"{ts} -> {file_mtime}")
    # Check if file date matches target date
    if (file_mtime == time.date()):
        matching_files.append(file)
```
